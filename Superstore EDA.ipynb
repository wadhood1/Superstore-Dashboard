```python
# Convert the column Order Date to datetime format
data['Order Date'] = pd.to_datetime(data['Order Date'], errors='coerce')

# Check for missing or incorrect dates
invalid_dates = data[data['Order Date'].isnull()]['Order Date']

# Define the two possible date formats
date_formats = ['%d/%m/%Y', '%m/%d/%Y']

# Iterate through the possible date formats
for date_format in date_formats:
    # Try to convert the dates using the current format
    data['Order Date'] = pd.to_datetime(data['Order Date'], format=date_format, errors='coerce')

    # Check for missing or incorrect dates after the conversion
    invalid_dates = data[data['Order Date'].isnull()]['Order Date']

    # If there are no invalid dates, break out of the loop
    if invalid_dates.empty:
        break

# Check again for invalid dates after attempting both formats
invalid_dates = data[data['Order Date'].isnull()]['Order Date']

# Print the DataFrame to see the results
data
```

```python
subcategory_counts = data['Sub-Category'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(subcategory_counts, labels=subcategory_counts.index,autopct='%1.1f%%')

# Add a title
plt.title('Sub-Category Distribution')

# Display the pie chart
plt.show()
```

```python
data.groupby('Sub-Category')[['Sales','Profit']].agg(['sum']).plot.bar()
plt.title('Total Sales and Profit by Sub-Category')
plt.figure.figsize = (10,8)
plt.show()
```

```python
data.groupby('Region')[['Sales','Profit']].agg(['sum']).plot.bar()
plt.title('Total Sales and Profit by Region')
plt.figure.figsize = (10,8)
plt.show()
```

```python
correlation_matrix = data.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
covariance_matrix = data.cov()

plt.figure(figsize=(10, 8))
sns.heatmap(covariance_matrix, annot=True, fmt=".2f")
plt.title('Covariance Heatmap')
plt.show()
```

```python
subcategory_profit = data.groupby('Sub-Category')['Profit'].mean()

plt.figure(figsize=(10, 6))
plt.plot(subcategory_profit.index, subcategory_profit.values, marker='o')

plt.xlabel('Sub-Category')
plt.ylabel('Profit')
plt.title('Sub-Category vs Profit')
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='Discount', y='Profit', label = 'Discount')


plt.title('Discount vs Profit')

# Show the plot
plt.show()
```
